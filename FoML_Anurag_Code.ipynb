## The following implementation of Arima gives us result of multiple Arima models, and hence we conclude which one is the best one and then report results for that and graph that results

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima
import matplotlib.pyplot as plt

# Load and preprocess the data
data = pd.read_excel('sample.xlsx', sheet_name='GlobalLandTemperaturesByCity')
new_york_data = data[data['City'] == 'Berlin']
new_york_data['dt'] = pd.to_datetime(new_york_data['dt'])
new_york_data.set_index('dt', inplace=True)
new_york_data = new_york_data['AverageTemperature'].resample('MS').mean()  # Monthly averages
new_york_data = new_york_data.ffill()  # Fill missing values using forward fill

# Ensure there are no empty data series
if new_york_data.empty:
    raise ValueError("No data available for New York. Check the data filtering and loading steps.")

# Check for stationarity
result = adfuller(new_york_data.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
if result[1] > 0.05:
    print("Series is not stationary. Differencing will be required.")
else:
    print("Series is stationary.")

# Fit an ARIMA model automatically
model = auto_arima(new_york_data, seasonal=False, stepwise=True, suppress_warnings=True, error_action="ignore", max_order=None, trace=True)
print(model.summary())

# Forecast
forecast = model.predict(n_periods=60)

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot(new_york_data, label='Historical Average Temperature')
plt.plot(pd.date_range(new_york_data.index[-1], periods=60, freq='MS'), forecast, color='red', label='Forecasted Temperature')
plt.title('Temperature Forecast for New York')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

### following is the test train and prediction analysis

```python
import pandas as pd

# Assuming 'combined_data' is your DataFrame and 'AverageTemperature' is your column of interest
# Calculate the split point
split_point = int(len(combined_data) * 0.8)
train, test = combined_data.iloc[:split_point], combined_data.iloc[split_point:]

# Now train the ARIMA model on the training set
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train['AverageTemperature'], order=(34,4,28))
fitted_model = model.fit()
print(fitted_model.summary())
# Forecast on the test set
forecast = fitted_model.get_forecast(steps=len(test))
forecasted_values = forecast.predicted_mean

# Comparing forecasted values against the actual test set
plt.figure(figsize=(10, 5))
plt.plot(train.index, train['AverageTemperature'], label='Training Data')
plt.plot(test.index, test['AverageTemperature'], label='Actual Test Data')
plt.plot(test.index, forecasted_values, label='Forecasted Test Data', color='red')
plt.title('ARIMA Model Train and Test Data')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

### MSE MAE and other Stats Reported
